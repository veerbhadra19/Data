```python
#Data cleansing
```

```python
df=data.replace('-1',np.NaN)
```

```python
df.head(20)
```

```python
new_df=df.fillna(method='ffill')
```

```python
new_df.head(20)
```

```python
new_df.replace({'prep_time':-1,'cook_time':-1},{'prep_time':20,'cook_time':30})
```

```python
plt.xlabel('preparation_time')
plt.ylabel('cook_time')
plt.title('Time Comparision')
plt.plot(new_df.prep_time,new_df.cook_time, color= 'red', linestyle ='dashed')
plt.grid()
```

```python
plt.xlabel('preparation_time')
plt.ylabel('cook_time')
plt.title('Time Comparision')
plt.plot(new_df.prep_time,new_df.cook_time, 'bo')
```

```python
plt.figure (figsize=(15,8))
plt.xlabel('Region')
plt.ylabel('State')
plt.title('State_Region Comparision')
plt.bar(new_df.region,new_df.state,width=0.4)
```

```python
plt.xlabel('Course')
plt.ylabel('Range')
plt.title('Course Items')
plt.hist(new_df.course, bins=3,rwidth=0.95, color='y')
```
